```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
pd.options.display.max_columns = None
```

```python
files = ['ACS_2011', 'ACS_2012', 'ACS_2013', 'ACS_2014', 'ACS_2015', 'ACS_2016', 'ACS_2017', 'ACS_2018']
```

```python
# creating
data_ = pd.DataFrame()
for i in range(len(files)):
    acs_data = pd.read_csv(files[i]+'.csv')
    acs_data['year'] = int(files[i][4:])
    data_ = data_.append(acs_data)
```

```python
list(data_.columns)
```

```python
zil_ind = pd.read_csv('Zip_Zri_MultiFamilyResidenceRental.csv')
zil_ind
```

```python
# fill zip codes with less than 5 digits - acs, geo_id
data_.geo_id = data_.geo_id.str.zfill(5)
```

```python
# fill zip codes with less than 5 digits - zil_ind, RegionName
zil_ind.RegionName = zil_ind.RegionName.str.zfill(5)
```

```python
cols_replace = list(zil_ind.columns)[7:]
cols_keep = list(zil_ind.columns)[:7]
```

```python
zil_new = zil_ind.melt(id_vars=cols_keep, value_vars = cols_replace)
```

```python
zil_new['variable']= pd.to_datetime(zil_new['variable'])
```

```python
zil_new['year'] = zil_new['variable'].apply(lambda a: a.year)
```

```python
zil_new.info()
```

```python
zil_new[zil_new['RegionName']=='99701']
```

```python
zil_yr = zil_new.groupby(['RegionName', 'year']).mean().reset_index()
zil_yr
```

```python
zil_yr.drop('SizeRank', inplace=True, axis=1)
```

```python
zil_yr.drop('RegionID', inplace=True, axis=1)
```

```python
zil_yr.info()
```

```python
# dropping na values and 2020 and 2019 values
zil_yr = zil_yr.dropna()
zil_final = zil_yr[zil_yr['year'] < 2019]
```

```python
df_final = zil_final.merge(data_, how='left', left_on = ['RegionName','year'], right_on=['geo_id', 'year'])
```
